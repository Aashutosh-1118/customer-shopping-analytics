```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python

# Checking if missing data or null values are present in the dataset

df.isnull().sum()
```

```python

# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python

# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
import pandas as pd

q1 = pd.read_sql("""
    SELECT gender, SUM(purchase_amount) AS revenue
    FROM customer
    GROUP BY gender
""", engine)

q2 = pd.read_sql("""
    SELECT customer_id, purchase_amount
    FROM customer
    WHERE discount_applied = 'Yes' AND purchase_amount >= (SELECT AVG(purchase_amount) FROM customer)
""", engine)

q3 = pd.read_sql("""
    SELECT item_purchased, ROUND(AVG(review_rating), 2) AS average_product_rating
    FROM customer
    GROUP BY item_purchased
    ORDER BY AVG(review_rating) DESC
    LIMIT 5
""", engine)

q4 = pd.read_sql("""
    SELECT shipping_type, ROUND(AVG(purchase_amount), 2) AS avg_purchase_amount
    FROM customer
    WHERE shipping_type IN ('Standard', 'Express')
    GROUP BY shipping_type
""", engine)

q5 = pd.read_sql("""
    SELECT subscription_status,
           COUNT(customer_id) AS total_customers,
           ROUND(AVG(purchase_amount), 2) AS avg_spend,
           ROUND(SUM(purchase_amount), 2) AS total_revenue
    FROM customer
    GROUP BY subscription_status
    ORDER BY total_revenue DESC, avg_spend DESC
""", engine)

q6 = pd.read_sql("""
    SELECT item_purchased,
           ROUND(100.0 * SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS discount_rate
    FROM customer
    GROUP BY item_purchased
    ORDER BY discount_rate DESC
    LIMIT 5
""", engine)

q7 = pd.read_sql("""
    WITH customer_type AS (
        SELECT customer_id, previous_purchases,
            CASE
                WHEN previous_purchases = 1 THEN 'New'
                WHEN previous_purchases BETWEEN 2 AND 10 THEN 'Returning'
                ELSE 'Loyal'
            END AS customer_segment
        FROM customer
    )
    SELECT customer_segment, COUNT(*) AS number_of_customers
    FROM customer_type
    GROUP BY customer_segment
""", engine)

q8 = pd.read_sql("""
    WITH item_counts AS (
        SELECT category, item_purchased,
               COUNT(customer_id) AS total_orders,
               ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(customer_id) DESC) AS item_rank
        FROM customer
        GROUP BY category, item_purchased
    )
    SELECT item_rank, category, item_purchased, total_orders
    FROM item_counts
    WHERE item_rank <= 3
""", engine)

q9 = pd.read_sql("""
    SELECT subscription_status, COUNT(customer_id) AS repeat_buyers
    FROM customer
    WHERE previous_purchases > 5
    GROUP BY subscription_status
""", engine)

q10 = pd.read_sql("""
    SELECT age_group, SUM(purchase_amount) AS total_revenue
    FROM customer
    GROUP BY age_group
    ORDER BY total_revenue DESC
""", engine)

print("All 10 queries executed.")
```

```python
print("Q1:\n", q1, "\n")
print("Q2 (sample):\n", q2.head(), "\n")
print("Q3:\n", q3, "\n")
print("Q4:\n", q4, "\n")
print("Q5:\n", q5, "\n")
print("Q6:\n", q6, "\n")
print("Q7:\n", q7, "\n")
print("Q8:\n", q8, "\n")
print("Q9:\n", q9, "\n")
print("Q10:\n", q10, "\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Chart 1 — Revenue by gender (Q1)
plt.figure(figsize=(6,4))
sns.barplot(data=q1, x='gender', y='revenue')
plt.title('Total Revenue by Gender')
plt.ylabel('Revenue ($)')
plt.show()

# Chart 2 — Customer segments (Q7)
plt.figure(figsize=(6,4))
sns.barplot(data=q7, x='customer_segment', y='number_of_customers')
plt.title('Customer Segments: New vs Returning vs Loyal')
plt.ylabel('Number of Customers')
plt.show()

# Chart 3 — Revenue by age group (Q10)
plt.figure(figsize=(7,4))
sns.barplot(data=q10, x='age_group', y='total_revenue')
plt.title('Revenue Contribution by Age Group')
plt.ylabel('Revenue ($)')
plt.show()

# Chart 4 — Top 5 products by discount rate (Q6)
plt.figure(figsize=(8,4))
sns.barplot(data=q6, x='discount_rate', y='item_purchased', orient='h')
plt.title('Top 5 Products by Discount Rate (%)')
plt.xlabel('Discount Rate (%)')
plt.show()

# Chart 5 — Avg spend: subscribers vs non-subscribers (Q5)
plt.figure(figsize=(6,4))
sns.barplot(data=q5, x='subscription_status', y='avg_spend')
plt.title('Average Spend: Subscribers vs Non-Subscribers')
plt.ylabel('Average Spend ($)')
plt.show()
```

```python
# Chart 6 — Top 5 most-rated products (Q3)
plt.figure(figsize=(8,4))
sns.barplot(data=q3, x='average_product_rating', y='item_purchased', orient='h', palette='viridis')
plt.title('Top 5 Products by Average Review Rating')
plt.xlabel('Average Rating')
plt.xlim(0, 5)
plt.show()

# Chart 7 — Shipping type comparison (Q4)
plt.figure(figsize=(5,4))
sns.barplot(data=q4, x='shipping_type', y='avg_purchase_amount', palette='Set2')
plt.title('Average Purchase Amount: Standard vs Express Shipping')
plt.ylabel('Avg Purchase Amount ($)')
plt.show()

# Chart 8 — Top 3 products per category (Q8) — grouped bar
plt.figure(figsize=(10,5))
sns.barplot(data=q8, x='item_purchased', y='total_orders', hue='category')
plt.title('Top 3 Most Purchased Products per Category')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Total Orders')
plt.legend(title='Category', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Chart 9 — Repeat buyers' subscription status (Q9) — donut/pie
plt.figure(figsize=(5,5))
plt.pie(q9['repeat_buyers'], labels=q9['subscription_status'], autopct='%1.1f%%',
        colors=sns.color_palette('pastel'), wedgeprops={'edgecolor':'white'})
plt.title('Subscription Status Among Repeat Buyers (>5 purchases)')
plt.show()

# Chart 10 — Age distribution overall (uses raw df, not a query result)
plt.figure(figsize=(7,4))
sns.histplot(df['age'], bins=20, kde=True, color='steelblue')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.show()

# Chart 11 — Purchase amount distribution by category (uses raw df)
plt.figure(figsize=(9,5))
sns.boxplot(data=df, x='category', y='purchase_amount', palette='coolwarm')
plt.title('Purchase Amount Spread by Category')
plt.xticks(rotation=30)
plt.show()

# Chart 12 — Correlation heatmap of numeric features (uses raw df)
plt.figure(figsize=(6,5))
numeric_cols = df[['age', 'purchase_amount', 'review_rating', 'previous_purchases', 'purchase_frequency_days']]
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Numeric Features')
plt.tight_layout()
plt.show()
```
